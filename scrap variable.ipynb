```python
# Column names
names = []
for v in all_values:
#     print(v.text)
    if v.text.startswith("\xa0\xa0"):
#         print(v.text[5:])
        name = v.text[5:]
    else:
#         print(v.text)
        name = v.text
    names.append(name)
print(names)
print(len(names))
```

```python
dfs = []
for i,name in enumerate(names):
    data= pd.read_excel(f"Data_Northern_Disease_excel/zone_summary_report({i+1}).xlsx",header = 30, skipfooter=2)
    df = data.iloc[:,:2]
    df.columns = ["Date",name]
#     print(df)
#     df[name] = df[name].str.replace("-","")
    dfs.append(df[name])
df1 = pd.concat(dfs,axis = 1)
# df1
```

```python
from datetime import datetime
# start = "2010-01-01"
df1["Date"] = pd.DataFrame({"Date":pd.date_range(start = datetime(2010,1,1),end = datetime(2022,12,31),freq ="M").tolist()})

df1.set_index("Date",inplace = True)
```

```python
import numpy as np
for column in df1.columns:
    df1[column] = df1[column].replace({"-":None})
    df1[column] = (df1[column].replace({"":None}))
df1                     
```

```python
df1.to_excel("Data_Disease_Northen.xlsx",index = True , encoding = "utf-8-sig")
```
